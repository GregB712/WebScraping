```python
# Import the libraries.
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import re
```

```python
# Extract the HTML and create a BeautifulSoup object.
url = ('https://www.tripadvisor.in/Hotels-g189473-Thessaloniki_Thessaloniki_Region_Central_Macedonia-Hotels.html')

user_agent = ({'User-Agent':
			'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
			AppleWebKit/537.36 (KHTML, like Gecko) \
			Chrome/90.0.4430.212 Safari/537.36',
			'Accept-Language': 'en-US, en;q=0.5'})

def get_page_contents(url):
    page = requests.get(url, headers = user_agent)
    return BeautifulSoup(page.text, 'html.parser')

soup = get_page_contents(url)

with open("home_page.html", "w") as file:
    file.write(str(soup))
```

```python
# Find and extract data elements.
ratings = []
for rating in soup.findAll('a',{'class':'ui_bubble_rating'}):
    ratings.append(rating['alt'])
reviews = []
for review in soup.findAll('a',{'class':'review_count'}):
    reviews.append(review.text.strip())
base_UrlString = 'https://www.tripadvisor.com'
hotels = []
links = []
for link in soup.findAll('a',{'class':'property_title prominent'}):
    links.append(base_UrlString + link['href'])
    hotels.append(re.sub(r'\d+\. ', '', link.text.strip()))

# Create the dictionary.
dict = {'Hotel Names':hotels,'Ratings':ratings,'Number of Reviews':reviews,'Link':links}

# Create the dataframe.
df = pd.DataFrame.from_dict(dict)
df = df.drop_duplicates().reset_index(drop=True)

# Define the user_agent
user_agent = ({'User-Agent':
			'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
			AppleWebKit/537.36 (KHTML, like Gecko) \
			Chrome/90.0.4430.212 Safari/537.36',
			'Accept-Language': 'en-US, en;q=0.5'})

# Apply the get_page_contents function to each line of the "Link" column
df['Page Contents'] = df['Link'].apply(get_page_contents)
```

```python
def getReviewInfo(business, url):
    print(business)
    current_page = 0
    increment = 10
    users = []
    dates = []
    visitDates = []
    titles = []
    reviews = []
    ratings = []
    businessList = []

    while True:
        if current_page == 0:
            new_url = url        
        else:
            new_url = url.replace("Reviews-", f"Reviews-or{current_page}-")
        current_page += increment
        test_soup = get_page_contents(new_url)
        reviewDiv = test_soup.findAll('div',{'class':'YibKl MC R2 Gi z Z BB pBbQr'})

        if len(reviewDiv) > 0:
            user_dateDiv = test_soup.findAll('div',{'class':'cRVSd'})
            visitDateSpan = test_soup.findAll('span',{'class':'teHYY _R Me S4 H3'})
            titleA = test_soup.findAll('a',{'class':'Qwuub'})
            reviewSpan = test_soup.findAll('span',{'class':'QewHA H4 _a'})
            ratingDiv = test_soup.findAll('div', {'class': 'Hlmiy F1'})

            print(new_url)
            for user_date, visitDate, title, review, rating in zip(user_dateDiv, visitDateSpan, titleA, reviewSpan, ratingDiv):
                # print(user_date)
                user_date = user_date.text.split(' wrote a review ')
                user = user_date[0]
                date = user_date[1]
                visitDate = visitDate.text.replace('Date of stay: ', '')
                title = title.text
                review = review.text
                rating = rating.find('span', {'class': 'ui_bubble_rating'})['class'][1]

                users.append(user)
                dates.append(date)
                visitDates.append(visitDate)
                titles.append(title)
                reviews.append(review)
                ratings.append(rating)
                businessList.append(business)
        else:
            break
        
    # Create the dictionary.
    dict = {'Business reviewed':businessList, 'Reviewer Username':users, 'Review date':dates, 
            'Visit Date':visitDates, 'Review Title':titles, 'Review Text':reviews, 'Review Ratings':ratings}

    # Create the dataframe.
    df = pd.DataFrame.from_dict(dict)
    return df
```

```python
df
```

```python
# Apply the getReviewInfo function to each row of the dataframe
resulting_dfs = df.apply(lambda row: getReviewInfo(row['Hotel Names'], row['Link']), axis=1)

# Concatenate the resulting dataframes
concatenated_df = pd.concat(resulting_dfs.values, ignore_index=True)

# Print the concatenated dataframe
print(concatenated_df)
```

```python
concatenated_df.to_csv('database.csv', index=False)
```
